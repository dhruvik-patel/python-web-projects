```python
df.drop('id',axis=1, inplace=True)
```

```python
df.time = pd.to_datetime(df.time)
```

```python
df.index = df.time
```

```python
plt.plot(df['feature_1'])
```

```python
plt.plot(df['feature_2'])
```

```python
df = df[['feature_1', 'feature_2']]
split = int(0.9*len(df))
train, test = df[0:split], df[split:]
```

```python
from statsmodels.tsa.statespace.varmax import VARMAX

model = VARMAX(train)
md_fit = model.fit()
```

```python
pred = md_fit.forecast(steps=len(test))
pred
```

```python
##################################################################################################
from sklearn import metrics
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(test, pred)))
print('Mean Absolute Error : ', metrics.mean_absolute_error(test, pred))
```

```python
test_arr = np.array(test.feature_2, dtype=float)
pred_arr = np.array(pred.feature_2, dtype=float)
```

```python
temp = pd.DataFrame({'Actual': test_arr.flatten(), 'Predicted': pred_arr.flatten()})
temp
```
