```python
X = X.drop(['Alley','FireplaceQu','PoolQC','Fence','MiscFeature','Id'],axis=1)
```

```python
####################  To Select Columns
# With pandas,
# wit column names
# dataframe[['column1','column2']]
# to select by iloc and specific columns with index number:
# dataframe.iloc[:,[1,2]]
# with loc column names can be used like
# dataframe.loc[:,['column1','column2']]
```

```python
#5 taking care of missing data for int

from sklearn.impute import SimpleImputer
simpleimp = SimpleImputer(missing_values=np.nan, strategy='mean')
simpleimp = simpleimp.fit(X.iloc[:,[2,24,56]])
X.iloc[:,[2,24,56]] = simpleimp.transform(X.iloc[:,[2,24,56]])
```

```python
#5 taking care of missing data for string

from sklearn.impute import SimpleImputer
simpleimp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
simpleimp = simpleimp.fit(X.iloc[:,[23,28,29,30,31,33,40,55,57,60,61]])
X.iloc[:,[23,28,29,30,31,33,40,55,57,60,61]] = simpleimp.transform(X.iloc[:,[23,28,29,30,31,33,40,55,57,60,61]])
```

```python
# encoding categorical data
# independent variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
onehotencoder = OneHotEncoder(categories='auto')
X = onehotencoder.fit_transform(X).toarray()
#dependent variables
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```
